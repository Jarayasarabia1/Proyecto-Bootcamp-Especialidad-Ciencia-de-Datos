```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split


df = pd.read_csv("dataset_natalidad.csv")

print("\n Dataset original: \n")
print(df)

print("\n 5 primeras filas del dataset: \n")
print(df.head())

print("\n Información general del dataset: \n")
print(df.info())

print("\n Resumen estadístico: \n")
print(df.describe())

df.dtypes

print(("\n Valores Nulos: \n"))
print(df.isnull())

print(("\n Suma valores nulos: \n"))
print(df.isnull().sum())

print("\n No se encontraron valores nulos en el dataset. \n")


#Analiza las correlaciones entre variables y visualiza su distribución.

# Correlación entre variables
corr = df.drop('País', axis=1).corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title("Correlación entre variables")
plt.show()

# Visualización de la distribución de las variables
df.hist(figsize=(12, 10), bins=15)
plt.suptitle("Distribución de variables socioeconómicas y tasa de natalidad")
plt.show()

#2). Diseño y entrenamiento del modelo.

# Definimos la red neuronal.

# Variables predictoras y objetivo
X = df.drop(columns=["Tasa_Natalidad", "País"])
y = df["Tasa_Natalidad"]

# Normalizar los datos
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)


# Crear el modelo de red neuronal
model = Sequential()

# Capa de entrada
model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))

# Capas ocultas
model.add(Dense(32, activation='relu'))
model.add(Dense(16, activation='relu'))

# Capa de salida con una neurona (
model.add(Dense(1))

# Recopilar el modelo con el optimizador Adam(learning_rate=0.001), loss='mean_squared_error')
model.compile(optimizer=Adam(learning_rate=0.001), loss='mean_squared_error')


# Resumen de la arquitectura del modelo
model.summary()

# Entrenar el modelo
history = model.fit(X_train, y_train, epochs=100, batch_size=8, validation_split=0.2)

# 3). Evaluación y optimización del modelo.

# Evaluar el modelo en datos de prueba
loss = model.evaluate(X_test, y_test)
print(f'Pérdida del modelo: {loss}')

# Graficar la pérdida de entrenamiento y validación
plt.plot(history.history['loss'], label='Pérdida de entrenamiento')
plt.plot(history.history['val_loss'], label='Pérdida de validación')
plt.title('Pérdida durante el entrenamiento')
plt.xlabel('Epochs')
plt.ylabel('Pérdida')
plt.legend()
plt.show()

# Realizar predicciones
y_pred = model.predict(X_test)

# Comparar las predicciones con los datos reales
plt.scatter(y_test, y_pred)
plt.xlabel('Tasa de natalidad real')
plt.ylabel('Predicción de la tasa de natalidad')
plt.title('Comparación de las predicciones con los valores reales')
plt.show()
```
